```python
import random
import numpy as np
import matplotlib.pyplot as plt

#Generate geometric random variables
def generate_geometric(p, size):
        random_variables = []  # Initialize an empty list to store generated random variables

        for _ in range(size):  # Loop 'size' times to generate the specified number of random variables
            trials = 1  # Initialize the number of trials to 1

        # Continue generating trials until a success occurs
        while random.random() > p:
            trials += 1  # Increment the number of trials

        random_variables.append(trials)  # Add the number of trials to the list

        return random_variables  # Return the list of generated random variables
    
# Calculate the PMF
def geometric_pmf(x, p):
    return (1 - p)**(x - 1) * p

# Calculate the CDF
def geometric_cdf(x, p):
    return 1 - (1 - p)**x
    
# Calculate the Expectation
def geometric_expectation(p):
    return 1 / p

# Calculate the Variance
def geometric_variance(p):
    return (1 - p) / (p ** 2)

# Set the probability of success
p = 0.3

# Set number of generated variables to 10000
size = 10000

#Store generated random variables
random_variables = generate_geometric(p, size)

# Get the unique values in generated geometric variables
unique_values, counts = np.unique(random_variables, return_counts=True)

#Store expectation and variance variables
expectation = geometric_expectation(p)
variance = geometric_variance(p)

#Print expectation and variance
print(f'Expectation (mean): {expectation:.2f}')
print(f'Variance: {variance:.2f}')


# Calculate the PMF values for each unique value
pmf_values = [geometric_pmf(x, p) for x in unique_values]

#Sort random variables to calculate CDF for each value
x_values = np.sort(random_variables)

#Calculate CDF by iterating through x_values
cdf_values = [geometric_cdf(x, p) for x in x_values]

#Plot PMF
plt.bar(unique_values, pmf_values, color='blue', alpha=0.7, label='PMF')
plt.title('Geometric Distribution PMF')
plt.xlabel('x (Number of trials until first success)')
plt.ylabel('Probability')
plt.legend()
plt.show()

#Plot CDF
plt.step(x_values, cdf_values, where='post', color='green', label='CDF')
plt.title('Geometric Distribution CDF')
plt.xlabel('x (Number of trials until first success)')
plt.ylabel('Cumulative Probability')
plt.legend()
plt.show()
```
