```python
import random
import numpy as np
import matplotlib.pyplot as plt

#Generate geometric random variables
def generate_geometric(p, size):
    random_variables = np.random.geometric(p=p, size=size)
    return random_variables
    
# Calculate the PMF
def geometric_pmf(x, p):
    return (1 - p)**(x - 1) * p

# Calculate the CDF
def geometric_cdf(x, p):
    return 1 - (1 - p)**x
    
# Calculate the Expectation
def geometric_expectation(p):
    return 1 / p

# Calculate the Variance
def geometric_variance(p):
    return (1 - p) / (p ** 2)

#Plot PMF
def plot_pmf(p):
    pmf_values = [geometric_pmf(x, p) for x in unique_values]
    plt.bar(unique_values, pmf_values, color='blue', alpha=0.7, width = 0.6, label='PMF')
    plt.scatter(unique_values, pmf_values, color='blue', marker='o', s=90)
    plt.title(f'Probability Mass Function for Geometric distribution (p={p:.2f})')
    plt.xlabel('x (Number of trials until first success)')
    plt.ylabel('Probability')
    plt.legend()
    plt.show()

#Plot CDF
def plot_cdf(p):
    cdf_values = [geometric_cdf(x, p) for x in unique_values]
    plt.step(unique_values, cdf_values, marker='o', where='post', color='green', label='CDF')
    plt.title(f'Cumulative Distributive Function for Geometric distribution (p={p:.2f})')
    plt.xlabel('x (Number of trials until first success)')
    plt.ylabel('Cumulative Probability')
    plt.legend()
    plt.show()

# Set the probability of success
p = 0.25

# Set number of generated variables to 10000
size = 1000

#Store generated random variables
random_variables = generate_geometric(p, size)

# Get the unique values in generated geometric variables
unique_values, counts = np.unique(random_variables, return_counts=True)

#print expectation and mean
print(f'Expectation (mean): {geometric_expectation(p):.2f}')
print(f'Variance: {geometric_variance(p):.2f}')

#show results
plot_pmf(p)
plot_cdf(p)
```
