```python
import numpy as np
import matplotlib.pyplot as plt

# Set the probability of success
p = 0.3

# Generate 1000 geometric random variables
random_variables = np.random.geometric(p, size=1000)

# Calculate the PMF using the mathematical formula
def geometric_pmf(x, p):
    return (1 - p)**(x - 1) * p

def geometric_cdf(x, p):
    return 1 - (1 - p)**x

def geometric_expectation(p):
    return 1 / p

def geometric_variance(p):
    return (1 - p) / (p ** 2)

# Count the occurrences of each value in the random variables
unique_values, counts = np.unique(random_variables, return_counts=True)

expectation = geometric_expectation(p)
variance = geometric_variance(p)
print(f'Expectation (mean): {expectation:.2f}')
print(f'Variance: {variance:.2f}')
# Calculate the PMF values for each unique value
pmf_values = [geometric_pmf(x, p) for x in unique_values]
x_values = np.sort(random_variables)
cdf_values = [geometric_cdf(x, p) for x in x_values]
plt.bar(unique_values, pmf_values, color='blue', alpha=0.7, label='PMF')
plt.title('Geometric Distribution PMF')
plt.xlabel('x (Number of trials until first success)')
plt.ylabel('Probability')
plt.legend()
plt.show()

plt.step(x_values, cdf_values, where='post', color='green', label='CDF')
plt.title('Geometric Distribution CDF')
plt.xlabel('x (Number of trials until first success)')
plt.ylabel('Cumulative Probability')
plt.legend()
plt.show()
```
