```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the continuous uniform distribution
a = int(input("Enter the lower limit : "))
b = int(input("Enter the higher limit : "))

# Define the PDF of the continuous uniform distribution
def continuous_uniform_pdf(x, a, b):
    return np.where((a <= x) & (x <= b), 1 / (b - a), 0)

# Define the CDF of the continuous uniform distribution
def continuous_uniform_cdf(x, a, b):
    return np.where(x < a, 0, np.where(x <= b, (x - a) / (b - a), 1))

# Variance of continuous uniform distribution
def continuous_uniform_variance(a, b):
    return ((b - a)**2) / 12

# Expectation (mean) of continuous uniform distribution
def continuous_uniform_expectation(a, b):
    return (a + b) / 2

def plot_continuous_uniform_pdf(a, b):
    # Generate a range of x values for plotting
    x_values = np.linspace(a-2, b+2, 1000)

    # Plot Probability Density Function (PDF)
    plt.plot(x_values, continuous_uniform_pdf(x_values, a, b), label='PDF')
    plt.title(f'PDF of U ~ ({a}, {b})')
    plt.xlabel('x')
    plt.ylabel('Density')
    plt.legend()
    plt.show()

def plot_continuous_uniform_cdf(a, b):
    # Generate a range of x values for plotting
    x_values = np.linspace(a-2, b+2, 1000)

    # Plot Cumulative Distribution Function (CDF) in green
    plt.plot(x_values, continuous_uniform_cdf(x_values, a, b), label='CDF', color='green')
    plt.title(f'CDF of U ~ ({a}, {b})')
    plt.xlabel('x')
    plt.ylabel('Probability')
    plt.legend()
    plt.show()

# Call the plotting functions
plot_continuous_uniform_pdf(a, b)
plot_continuous_uniform_cdf(a, b)

Variance = continuous_uniform_variance(a, b)
Expectation = continuous_uniform_expectation(a, b)

print("Variance is : ", round(Variance, 2))
print("Expectation is : ", round(Expectation, 2))
```
