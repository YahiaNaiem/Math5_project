```python
import numpy as np
import matplotlib.pyplot as plt
import math

# Generate Poisson random variables
def generate_poisson(lambda_val, size):
    poisson_values = np.random.poisson(lambda_val, size)
    return poisson_values

# Calculate pmf
def poisson_pmf(k, lambda_val):
    return (lambda_val**k * math.exp(-lambda_val)) / math.factorial(k)

# Calculate cdf
def poisson_cdf(k, lambda_val):
    cdf = 0.0
    for i in range(k + 1):
        cdf += poisson_pmf(i, lambda_val)
    return cdf

# Calculate expectation using the mathematical formula
def poisson_expectation(lambda_param):
    return lambda_param

# Calculate variance using the mathematical formula
def poisson_variance(lambda_param):
    return lambda_param

# Plot pmf
def plot_poisson_pmf(lambda_val, unique_values):
    pmf_values = [poisson_pmf(k, lambda_val) for k in unique_values]
    plt.scatter(unique_values, pmf_values, color='blue', marker='o', s=70)
    plt.bar(unique_values, pmf_values, color='blue', alpha=0.7, width=0.1, label='PMF')

    plt.title(f'Poisson PMF for  = {lambda_val}')
    plt.xlabel('k (Number of events)')
    plt.ylabel('Probability')
    plt.legend()
    plt.show()

# Plot cdf
def plot_poisson_cdf(lambda_val, unique_values):
    cdf_values = [poisson_cdf(k, lambda_val) for k in unique_values]

    plt.step(unique_values, cdf_values, where='post', color='green', marker='o', label='CDF')
    plt.title(f'Poisson CDF for  = {lambda_val}')
    plt.xlabel('k (Number of events)')
    plt.ylabel('Cumulative Probability')
    plt.legend()
    plt.show()

# Calculate overall average lambda
def calculate_overall_average_lambda(people):
    # Initialize an empty list to store average lambdas for each interval
    average_lambdas = []
    # Initialize an empty list to store the generated samples
    generated_samples = []
    # Iterate over each number of people
    for num_people in people:
        # Generate a random sample from a Poisson distribution with the given lambda
        sample = np.random.poisson(num_people, 1000)
        # Calculate the average lambda for the current interval
        average_lambda = np.mean(sample)
        # Append the average lambda to the list
        average_lambdas.append(average_lambda)
        # Append the generated sample to the list
        generated_samples.append(sample)
    # Calculate the overall average of lambdas
    overall_average_lambda = np.mean(average_lambdas)
    # Print the overall average
    print("Overall Average Lambda:", overall_average_lambda)
    # Return the generated samples and the overall average lambda
    return generated_samples, overall_average_lambda

# Example using the provided data
people = [
    24.231, 13.002, 17.139, 11.82, 8.274, 14.184,
    7.092, 26.004, 23.049, 43.143, 65.01, 47.28,
    27.186, 26.004, 31.914, 27.186, 23.049, 28.959,
    30.141, 17.73, 17.139, 28.959, 23.049, 11.229
]

generated_samples, overall_average_lambda = calculate_overall_average_lambda(people)

# Now you can use the generated samples and overall average lambda in the first code as needed.
# For example, you can plot the Poisson PMF and CDF for the first generated sample.
unique_values = np.unique(generated_samples[0])
plot_poisson_pmf(overall_average_lambda, unique_values)
plot_poisson_cdf(overall_average_lambda, unique_values)
print(f"Mean and variance = {overall_average_lambda}")
# Calculate overall average lambda
def calculate_overall_average_lambda(people):
    # Initialize an empty list to store average lambdas for each interval
    average_lambdas = []
    # Initialize an empty list to store the generated samples
    generated_samples = []
    # Iterate over each number of people
    for num_people in people:
        # Generate a random sample from a Poisson distribution with the given lambda
        sample = np.random.poisson(num_people, 1000)
        # Calculate the average lambda for the current interval
        average_lambda = np.mean(sample)
        # Append the average lambda to the list
        average_lambdas.append(average_lambda)
        # Append the generated sample to the list
        generated_samples.append(sample)
    # Calculate the overall average of lambdas
    overall_average_lambda = np.mean(average_lambdas)
    # Print the overall average
    print("Overall Average Lambda:", overall_average_lambda)
    # Return the generated samples and the overall average lambda
    return generated_samples, overall_average_lambda

# Example using the provided data
people = [
    24.231, 13.002, 17.139, 11.82, 8.274, 14.184,
    7.092, 26.004, 23.049, 43.143, 65.01, 47.28,
    27.186, 26.004, 31.914, 27.186, 23.049, 28.959,
    30.141, 17.73, 17.139, 28.959, 23.049, 11.229
]
```
