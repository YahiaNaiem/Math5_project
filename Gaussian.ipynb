```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Probability Density Function (PDF) for Gaussian distribution
def gaussian_pdf(x, mean, std_dev):
    return stats.norm.pdf(x, mean, std_dev)

# Cumulative Distribution Function (CDF) for Gaussian distribution
def gaussian_cdf(x, mean, std_dev):
    return stats.norm.cdf(x, mean, std_dev)

# Variance of Gaussian distribution
def gaussian_variance(std_dev):
    return std_dev**2

def gaussian_expectation(mu):
    # Calculate the expectation (mean) of a Gaussian distribution.
    return mu

def plot_gaussian_pdf(mean, std_dev, size=10000):
    # Generate a random sample from a Gaussian distribution
    X = np.random.normal(mean, std_dev, size=size)

    # Generate values for the PDF curve
    x_pdf = np.linspace(min(X), max(X), 1000)
    pdf_values = gaussian_pdf(x_pdf, mean, std_dev)

    # Plot Probability Density Function (PDF) as a curve and histogram
    plt.plot(x_pdf, pdf_values)
    #plt.hist(X, bins='auto', density=True, alpha=0.5, label='Empirical PDF (Histogram)')
    plt.xlabel('X')
    plt.ylabel('PDF')
    plt.title(f'Probability Density Function for Gaussian Distribution (mean={mean}, std_dev={std_dev})')
    plt.legend()
    plt.show()

def plot_gaussian_cdf(mean, std_dev, size=10000):
    # Generate a random sample from a Gaussian distribution
    X = np.random.normal(mean, std_dev, size=size)

    # Sort the sample for plotting the empirical CDF
    x_cdf = np.sort(X)
    y_cdf = [gaussian_cdf(x, mean, std_dev) for x in x_cdf]

    # Plot Cumulative Distribution Function (CDF)
    plt.step(x_cdf, y_cdf, where='post', color = "green")
    plt.title(f'Cumulative Distribution Function for Gaussian Distribution (mean={mean}, std_dev={std_dev})')
    plt.xlabel('X')
    plt.ylabel('CDF')
    plt.legend()
    plt.show()

# Parameters for the Gaussian distribution
mean = 3
std_dev = 1.50

# Call the plotting functions
plot_gaussian_pdf(mean, std_dev)
plot_gaussian_cdf(mean, std_dev)
```
